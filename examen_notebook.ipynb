### Explicaci√≥n del Ejercicio 2

**¬øC√≥mo funciona?**
1. Crea matriz resultado C de n√ón inicializada en ceros
2. Triple bucle anidado:
   - i: recorre filas de A
   - j: recorre columnas de B
   - k: acumula productos A[i][k] √ó B[k][j]
3. Cada C[i][j] = suma de productos punto entre fila i de A y columna j de B

**Recursos matem√°ticos:**
- Producto matricial: C[i][j] = Œ£(k=0 to n-1) A[i][k] √ó B[k][j]
- √Ålgebra lineal: el elemento (i,j) es el producto punto de fila i con columna j

**Estructuras de programaci√≥n:**
- Tres bucles `for` anidados
- List comprehension para inicializaci√≥n de matriz
- Acceso indexado bidimensional

**¬øPor qu√© es √≥ptima?**
- O(n¬≥) es el l√≠mite inferior te√≥rico para multiplicaci√≥n matricial directa
- Strassen (O(n^2.807)) tiene constantes muy altas, solo conviene para n > 1000
- Prec√°lculo de `fila_a = A[i]` mejora localidad de cach√©
